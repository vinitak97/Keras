# Preprocess data for training with Keras

```python
import numpy as np
from random import randint
from sklearn.preprocessing import MinMaxScaler
```

```python
train_labels = []
train_samples = []
```

Example data:
- An experimental drug was tested on individuals from ages 13 to 100.
- The trial had 2100 participants. Half were under 65 years old, half were over 65 years old.
- 95% of patients 65 or older experienced side effects.
- 95% of patients under 65 experienced no side effects.

```python
for i in range(1000):
    random_younger = randint(13,64)
    train_samples.append(random_younger)
    train_labels.append(0)
    
    random_older = randint(65,100)
    train_samples.append(random_older)
    train_labels.append(1)

for i in range(50):
    random_younger = randint(13,64)
    train_samples.append(random_younger)
    train_labels.append(1)
    
    random_older = randint(65,100)
    train_samples.append(random_older)
    train_labels.append(0)  
```

```python
#print raw data
for i in train_samples:
    print(i)
```

```python
for i in train_labels:
    print(i)
```

```python
train_labels = np.array(train_labels)
train_samples = np.array(train_samples)
```

```python
scaler = MinMaxScaler(feature_range=(0,1))
scaled_train_samples = scaler.fit_transform((train_samples).reshape(-1,1))
```

```python
for i in scaled_train_samples:
    print(i)
```

# ============================================================

 - <font color=red>**Keras**</font> is an open-source neural-network library.
 *** 
 - You should import the <font color=red>**backend**</font> module (tensorflow) as Keras cannot handle low-level operations such as tensor products, convolutions and so on itself. Instead, it relies on a specialized, well optimized tensor manipulation library to do so, serving as the <font color=blue>*"backend engine"*</font> of Keras.
 ***
 - The <font color=red>**Sequential model**</font> is a linear stack of layers (layers of neural network). You can create a Sequential model by passing a list of layer instances to the constructor.
 ***
 - <font color=red>**Activation functions**</font> are meant to learn and make sense of complicated and Non-linear complex functional mappings between the inputs and response variable.
 
 - A Neural Network without Activation function would simply be a <font color=blue>*Linear regression Model*</font>, which has limited power and does not perform good most of the times. Activation functions make *backpropagations* possible.
 
 - Activation function decides, *whether a neuron should be activated or not by calculating weighted sum and further adding bias with it*. The purpose of the activation function is to introduce *non-linearity into the output of a neuron*.
 ***
 - <font color=red>**ReLU**</font> is <font color=blue>Rectified Linear Unit activation function</font>, defined as <font color=green>y = max(0, x)</font>. The non-linear function is zero for negative values, and it grows linearly for positive values which is used only for hidden layers.
 *** 
 - The <font color=red>**softmax function**</font> is a more generalized logistic activation function which is used for <font color=blue>multiclass classification</font> for output layer. In mathematical definition way of saying the sigmoid function take any range real number and returns the output value which falls in the range of 0 to 1.
 ***
 - A <font color=red>**dense layer**</font> represents a <font color=blue>matrix vector multiplication</font>. 
 - Dense implements the operation: <font color=blue>*output = activation(dot(input, kernel) + bias)*</font> where activation is the element-wise activation function passed as the activation argument, kernel is a weights matrix created by the layer, and bias is a bias vector created by the layer.
 ***
 - An <font color=red>**optimizer**</font> is one of the two arguments required for compiling a Keras model. The Adam optimization algorithm is an extension to stochastic gradient descent to update network weights iterative based in training data.
 ***
 - When you are using a multiclass problem, you have to use <font color=red>**categorical_crossentropy**</font>. Here your targets should be in categorical format (e.g. if you have 10 classes, the target for each sample should be a 10-dimensional vector)
 


# ===========================================================

```python
import keras
from keras import backend as k
from keras.models import Sequential
from keras.layers import Activation
from keras.layers.core import Dense
from keras.optimizers import Adam
from keras.metrics import categorical_crossentropy
```

# Create an artificial neural network with Keras

```python
model = Sequential([
    Dense(16, input_shape=(1,), activation="relu"),
    Dense(32, activation="relu"),
    Dense(2, activation="softmax")
])
```

```python
model.summary()
```

# Train an artificial neural network with Keras

```python
model.compile(Adam(lr=0.0001), loss='sparse_categorical_crossentropy', metrics=['accuracy'])
```

```python
model.fit(scaled_train_samples, train_labels, batch_size=10, epochs=20, shuffle=True, verbose=2)
```

# Preprocess the Test Data 

```python
test_labels = []
test_samples = []
```

```python
for i in range(10):
    # The 5% of younger individuals who did experience side effects
    random_younger = randint(13,64)
    test_samples.append(random_younger)
    test_labels.append(1)
    
    # The 5% of older individuals who did not experience side effects
    random_older = randint(65,100)
    test_samples.append(random_older)
    test_labels.append(0)

for i in range(200):
    # The 95% of younger individuals who did not experience side effects
    random_younger = randint(13,64)
    test_samples.append(random_younger)
    test_labels.append(0)
    
    # The 5% of older individuals who did experience side effects
    random_older = randint(65,100)
    test_samples.append(random_older)
    test_labels.append(1)
```

```python
test_labels = np.array(train_labels)
test_samples = np.array(train_samples)
```

```python
scaler = MinMaxScaler(feature_range=(0,1))
scaled_test_samples = scaler.fit_transform((test_samples).reshape(-1,1))
```

# Predictions on Test Data

```python
predictions = model.predict(scaled_test_samples, batch_size=10, verbose=0)
```

```python
for i in predictions:
    print(i)
```

```python
rounded_predictions = model.predict_classes(scaled_test_samples, batch_size=10, verbose=0)
```

```python
for i in rounded_predictions:
    print(i)
```

# Creating a Confusion Matrix to analyze our prediction

```python
%matplotlib inline
from sklearn.metrics import confusion_matrix
import itertools
import matplotlib.pyplot as plt
```

```python
cm = confusion_matrix(test_labels, rounded_predictions)
```

```python
def plot_confusion_matrix(cm, classes, 
                          normalize=False, 
                          title='Confusion matrix', 
                          cmap=plt.cm.Blues):
    
    """
    This function prints and plots the confusion matrix.
    Normalization can be applied by setting 'normalize=True'
    """
    
    plt.imshow(cm, interpolation='nearest', cmap=cmap)
    plt.title(title)
    plt.colorbar()
    tick_marks = np.arange(len(classes))
    plt.xticks(tick_marks, classes, rotation=45)
    plt.yticks(tick_marks, classes)
    
    if normalize:
        cm = cm.astype('float')/cm.sum(axis=1)[:, np.newaxis]
        print("Normalized confusion matrix")
    else:
        print("Confusion matrix, without normalization")
    
    print(cm)
    
    thresh = cm.max()/2.0
    for i, j in itertools.product(range(cm.shape[0]), range(cm.shape[1])):
        plt.text(j, i, cm[i, j], horizontalalignment="center", color="white" if cm[i, j] > thresh else "black")
    plt.tight_layout()
    plt.ylabel("True Label")
    plt.xlabel("Predicted label")
```

```python
cm_plot_labels = ["No side effects", "Had side effects"]
plot_confusion_matrix(cm, cm_plot_labels, title="Confusion matrix")
```
